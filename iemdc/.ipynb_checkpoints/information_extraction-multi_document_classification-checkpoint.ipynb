### Reading the file data

```python
with open(file_path, 'r') as fp:
    data = fp.read()
```

### Preparing the data

```python
# Preprocessing
data = tp.strip_html_tags(data)
data = tp.remove_accented_chars(data)
data[:100]
```

```python
token_data = list()
roman_numbers_list = ['i.', 'ii.', 'iii.', 'iv.', 'v.', 'vi.', 'vii.', 'viii.', 'ix.', 'x.',
                      'i', 'ii', 'iii', 'iv', 'v', 'vi', 'vii', 'viii', 'ix', 'x']
punctuation_list = ['.', '[', ']', '-']
alphabets_list = list(string.ascii_lowercase)

unigram_word_test_list = roman_numbers_list + punctuation_list + alphabets_list

# iterate through each sentence in the file
for i in sent_tokenize(data):
    
    temp = []

    # tokenising based on spacy 'en_core_web_md' model
    # only for pre-processing
    # text preprocessing to lemmatize all the possible words
    le = tp.lemmatize_sp(i)
    # text preprocessing to remove stopwords
    sw = tp.remove_stopwords(le)
    # for further tokenization, we'll rely on word_tokenize. so let's join.
    i = ' '.join(sw)
    # also CAREFULLY, removing characters which are not required and doesn't hamper the accuracy
    i = i.replace(',', '').replace('"','').replace('(', ' ').replace(')', ' ').replace('/', ' ').replace(':', ' ').replace(';', ' ')
    i = re.sub('\d', ' ', i)

    # tokenize the sentence into words
    for j in word_tokenize(i):
        word = j.lower()
        if word not in unigram_word_test_list:
            temp.append(word)
    
    # adding the temp data to token data
    token_data.append(temp)
token_data[1]
```

### topic modeling

```python
bigram = gensim.models.Phrases(token_data, min_count=5, threshold=100) # higher threshold fewer phrases.
```

```python
id2word = gensim.corpora.Dictionary(token_data)
```

```python
corpora = [id2word.doc2bow(text) for text in token_data]
```

```python
print(corpora[15])
```

Gensim creates a unique id for each word in the document. The produced corpus shown above is a mapping of (word_id, word_frequency).

For example, (0, 1) above implies, word id 5 occurs twice in the first document. Likewise, word id 1 occurs once and so on.

```python
[[(id2word[id], freq) for id, freq in cp] for cp in corpora[15:16]]
```

```python
def convert_word2id(word):
    try:
        return list(id2word.keys())[list(id2word.values()).index(word)]
    except ValueError:
        print('Value not present')
        return None
```

```python
# Build LDA model
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpora,
                                           id2word=id2word,
                                           num_topics=50, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=300,
                                           passes=10,
                                           alpha='auto',
                                           per_word_topics=True)
```

```python
lda_model.get_term_topics(convert_word2id('majeure'))
```

```python
lda_model.get_term_topics(convert_word2id('notify'))
```

#### Computing the model perplexity and coherence score

```python
# Compute Perplexity
print('\nPerplexity: ', lda_model.log_perplexity(corpora))  # a measure of how good the model is. lower the better.

# Compute Coherence Score
coherence_model_lda = CoherenceModel(model=lda_model, texts=token_data, dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)
```

```python
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=3):
    """
    Compute c_v coherence for various number of topics

    Parameters:
    ----------
    dictionary : Gensim dictionary
    corpus : Gensim corpus
    texts : List of input texts
    limit : Max num of topics

    Returns:
    -------
    model_list : List of LDA topic models
    coherence_values : Coherence values corresponding to the LDA model with respective number of topics
    """
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        model = gensim.models.wrappers.LdaMallet(mallet_path, corpus=corpus, num_topics=num_topics, id2word=id2word)
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

    return model_list, coherence_values
```

```python
def format_topics_sentences(ldamodel, corpus, texts):
    # Init output
    sent_topics_df = pd.DataFrame()

    # Get main topic in each document
    for i, row in enumerate(ldamodel[corpus]):
        row = sorted(row, key=lambda x: (x[1]), reverse=True)
        # Get the Dominant topic, Perc Contribution and Keywords for each document
        for j, (topic_num, prop_topic) in enumerate(row):
            if j == 0:  # => dominant topic
                wp = ldamodel.show_topic(topic_num)
                topic_keywords = ", ".join([word for word, prop in wp])
                sent_topics_df = sent_topics_df.append(pd.Series([int(topic_num), round(prop_topic,4), topic_keywords]), ignore_index=True)
            else:
                break
    sent_topics_df.columns = ['Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords']

    # Add original text to the end of the output
    contents = pd.Series(texts)
    sent_topics_df = pd.concat([sent_topics_df, contents], axis=1)
    return(sent_topics_df)
```

```python
df_topic_sents_keywords = format_topics_sentences(ldamodel=lda_model, corpus=corpora, texts=token_data)
```
